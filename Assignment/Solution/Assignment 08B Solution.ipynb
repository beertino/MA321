**Question 8.4**

In *number theory*, a **perfect number** is a positive integer that is equal to the sum of its positive divisors, excluding the number itself. For instance, 6 has divisors 1, 2 and 3 (excluding itself), and 1 + 2 + 3 = 6, so 6 is a perfect number.

Print all the perfect numbers less than 10000. (There are 4 perfect numbers less than 10000)

````python
num = 1
keepgoing = True

while keepgoing:
    
    d = 1
    divisor_sum = 0
    keepgoing_D = (d < num)
    
    while keepgoing_D:
        
        if num % d == 0:
            divisor_sum = divisor_sum + d
        
        d = d + 1
        keepgoing_D = (d < num)
    
    if divisor_sum == num:
        print(num)
    
    num = num + 1
    keepgoing = (num < 10000)
````

In a number guessing game, the host picks a "secret number" between two values inclusive (e.g. 1 and 100), and asks the player to make a guess within the range in each turn. If the guess is wrong, the host will tell the player an updated range for the "secret number" based on the player's guess.

Write a program to simulate the following version of this game. 

In this game, the computer will act as a host and generate a "secret number" between 1 and 100. The player has up to 7 turns to guess this number. In each turn,
- display the current turn number, 
- display the current range,
- prompt the player to make a guess,
- if the guess is correct, display "Congratulations! and terminate the game.
- if the guess is incorrect, display "Wrong guess!" and update the upper or lower bound of the range.
- if the guess is not in the range provided, display "Out of range!" and "Wrong guess!", the upper and lower bounds remain unchanged.
- if the player fails to guess the number within 7 turns, display "You have used up the 7 guesses." and terminate the game.

````python
import random

def guess_number(l_bound, u_bound, secret, guess):
    
    if guess < l_bound or guess > u_bound:
        
        print("Out of Range!")
        return False, l_bound, u_bound
    
    if guess < secret:
        
        return False, guess + 1, u_bound
    
    if guess > secret:
        
        return False, l_bound, guess - 1
    
    return True, guess, guess

s = random.randint(1, 100)

turn = 1
l = 1
u = 100
keepgoing = True

while keepgoing:
    
    print("Turn No.", turn)
    print(f"The secret number is between {l} and {u} inclusive.")
    g = int(input("Your guess: "))
    
    result, l, u = guess_number(l, u, s, g)
    
    if result:
        print("Congratulations!")
        break
    else:
        print("Wrong guess!")
    
    turn = turn + 1
    keepgoing = (turn <= 7)

if turn == 8:
    print("You have used up the 7 guesses.")
````
